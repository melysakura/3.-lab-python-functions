## Functions

```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    
    for product in products:
        quantity = int(input("Please add the number of items for " + product))
        inventory[product] = quantity
    return inventory
```

```python
inventory = initialize_inventory(products)
```

```python
print(inventory)
```

```python
def get_customer_orders():
    customer_orders = set()
    order_again = "yes"
    
    while order_again == "yes":
        product = input("Please add the product you would like to order:").lower().strip()
        customer_orders.add(product)
        order_again = input("Would you like to continue ordering? Please answer with yes or no.").lower().strip()

    print ("Thank you for your order!")
    return customer_orders
```

```python
customer_orders = get_customer_orders()
```

```python
def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        inventory[product]-=1
    return inventory
```

```python
inventory = update_inventory(customer_orders, inventory)
```

```python
inventory
```

```python
def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_unique_products_ordered = (total_products_ordered / len(products))*100
    return total_products_ordered, percentage_unique_products_ordered
```

```python
order_statistics = calculate_order_statistics(customer_orders, products)
```

```python
order_statistics
```
